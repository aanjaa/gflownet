```python
import pandas
# import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
names = ["Random Action", "Random Trajectories", "Tempered Policy", "Temperature Conditioning", "Tempering + Random Action", "No Exploration"]
paths = ["logs/15650steps_20samples_exploration/seh_frag_e_random_action/dataframe.csv", "logs/15650steps_20samples_exploration/seh_frag_e_random_traj/dataframe.csv", "logs/15650steps_20samples_exploration/seh_frag_temp_fixed/dataframe.csv", "logs/15650steps_20samples_exploration/seh_frag_temp_cond/dataframe.csv", "logs/15650steps_20samples_exploration/seh_frag_temp_and_random_action/dataframe.csv", "logs/15650steps_20samples_exploration/seh_frag_no_exploration/dataframe.csv"]
```

```python
cols_to_plot = ["num_of_modes", "avg_topk", "overall_max_reward", "num_candidates_above_reward_thresh"]
col_to_plot = "val_avg_topk"
datas = [[] for _ in cols_to_plot]
data = []
for i, path in enumerate(paths):
    df = pandas.read_csv(path)
    data.append(df[col_to_plot].to_numpy())
    for j, col in enumerate(cols_to_plot):
        datas[j].append(df[col].to_numpy())
    # plt.plot(df[col_to_plot], label=path)
```

```python
data
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8, 10))

for i, ax in enumerate(axs.flat):
    ax.set_title(cols_to_plot[i])
    sns.violinplot(data=datas[i], ax=ax)
    # ax.violinplot(datas[i], showmeans=True)
    if i > 1:
        ax.set_xticks([0, 1, 2, 3, 4])
        ax.set_xticklabels(names)
    # ax.set_xlabel("Exploration Strategy")
    # ax.set_ylabel(cols_to_plot[i])
    plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right')
# axs.violinplot(data, showmeans=True)
# # axs.set_xticks([1, 2, 3, 4, 5])
# # axs.set_xticklabels(names)
# axs.set_xlabel("Exploration Strategy")
# axs.set_ylabel("Average Top-k Reward")
# plt.show()
```
